```python
#import necessary pacakages
import pandas as pd
import numpy as np
```

```python
#import women well being csv and create df
wwb_indicators_csv = "Women_well_being_clustered_data.csv"

wwb_indicators_df = pd.read_csv(wwb_indicators_csv)
wwb_indicators_df.head()
```

```python
#import csv for NaN lat long and gdp
missing_data_csv = "missing_data.csv"

missing_data_df = pd.read_csv(missing_data_csv)
missing_data_df.head()
```

```python
#list full column names
list(wwb_indicators_df.columns)
```

```python
#create new data frame with mean over each indicator, grouped by country
#overall_mean_RELSPEED = player_df_with_batter.groupby(["PITCH_TYPE"]).mean()["RELSPEED"]
DV_rate_per_country = wwb_indicators_df.groupby(["country_name"]).mean()["DV_phys_or_sex_partner_p"]
Mean_age_per_country = wwb_indicators_df.groupby(["country_name"]).mean()["DM_age_mean"]
Mean_age_when_married_per_country = wwb_indicators_df.groupby(["country_name"]).mean()["DM_age_marr_mean"]
Mean_ED_years_per_country = wwb_indicators_df.groupby(["country_name"]).mean()["ED_educ_years_mean"]
Fertility_rate_per_country = wwb_indicators_df.groupby(["country_name"]).mean()["FF_TFR"]
class_per_country = wwb_indicators_df.groupby(["country_name"]).mean()["class"]
lat_per_country = wwb_indicators_df.groupby(["country_name"]).mean()["latitude"]
long_per_country = wwb_indicators_df.groupby(["country_name"]).mean()["longitude"]
GDP_per_country = wwb_indicators_df.groupby(["country_name"]).mean()["GDP"]
```

```python
#create dataframe
indicators_per_country_df = pd.DataFrame([DV_rate_per_country, Mean_age_per_country, Mean_age_when_married_per_country, Mean_ED_years_per_country, Fertility_rate_per_country, class_per_country, lat_per_country, long_per_country, GDP_per_country])
#switch rows and columns aka transpose
indicators_per_country_df = indicators_per_country_df.T

indicators_per_country_df.head()
```
