```python
import pandas as pd
df = pd.DataFrame
```

```python
data_file = "Resources/livwell_lin_interpolated.csv"
data_file_df = pd.read_csv(data_file)
data_file_df
```

```python
dv_df = data_file_df.loc[:, data_file_df.columns.str.startswith('DV_')]
dv_df.columns
```

```python
dv_df = data_file_df[['DV_phys_or_sex_partner_p','country_name','year']]
dv_df
```

```python
edu_df = data_file_df.loc[:, data_file_df.columns.str.startswith('ED_')]
edu_df.columns
```

```python
edu_df = data_file_df[['ED_attainment_no_educ_p',
       'ED_attainment_primary_completed_p',
        'ED_attainment_secondary_completed_p',
       'ED_attainment_secondary_higher_p','country_name']]
edu_df
```

```python
dv_edu_df = data_file_df[['ED_attainment_no_educ_p',
       'ED_attainment_primary_completed_p',
        'ED_attainment_secondary_completed_p',
       'ED_attainment_secondary_higher_p','country_name', 'DV_phys_or_sex_partner_p','year']]
dv_edu_df
```

```python
lat_long_file = "Resources/country_lat_long.csv"
lat_long_df = pd.read_csv(lat_long_file)
lat_long_df
```

```python
count_lat_long = lat_long_df[['latitude','country','longitude']]
count_lat_long
```

```python
country_df = count_lat_long.rename(columns = {'country':'country_name'})
country_df
```

```python
merge_df = pd.merge(country_df,dv_edu_df,on="country_name")
merge_df
```

```python
merge_df = merge_df.dropna(how="any")
merge_df
```

```python
merge_df.to_csv('clean_data.csv', header=False, index=False)
```
